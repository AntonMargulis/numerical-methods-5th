```python
import math
pi=math.pi
arr_points = [0, pi / 6, pi * 2 / 6, pi * 3 / 6]
our_X = pi / 4
print(arr_points)
```

```python
from sympy import *
def lagranj(ourX, arr_points):
    x = symbols('x')
    a = []
    
    for number in range(len(arr_points)):
        sum = 1
        for i in range(len(arr_points)):
            if (i != number):
                sum *= (x - arr_points[i]) / (arr_points[number] - arr_points[i])   
        a.append(math.cos(arr_points[number]) * sum)
    sum = 0
    
    for number in range(len(a)):
        sum += a[number]
        
    return sum.subs({x:ourX})
```

```python
z = []
print("Значение в точке X* = ", our_X, ", полученное многочленом Лагранжа, равно ", lagranj(our_X, arr_points))
```

```python
x_lag = []
y_lag = []
x_cos = []
y_cos = []
for num in range(800):
    x_cos.append(-3 + num*0.01)
    y_cos.append(math.cos(-3 + num*0.01))
    x_lag.append(-3 +num * 0.01)
    y_lag.append(lagranj(-3 + num * 0.01, arr_points))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.title('Lagranj fuction')
plt.grid(True)
plt.scatter(math.pi / 4, lagranj(our_X, arr_points), color='red')
plt.plot(x_cos, y_cos, color='orange')
plt.plot(x_lag, y_lag, color='blue')
plt.show()
```

```python
#Ньютон
funct_n = []
def find_functions():
    
    for i in range(len(arr_points)):
        funct_n.append([])
        
    for i in range(len(arr_points)):
        funct_n[i].append(math.cos(arr_points[i]))
        
    for j in range(1, len(arr_points)):
        for i in range(len(arr_points)):
            if (i <= len(arr_points) - j - 1):
                funct_n[i].append((funct_n[i + 1][j - 1] - funct_n[i][j - 1]) / (arr_points[i + j] - arr_points[i]))
```

```python
def newton(f_of, ourX, arr_points):
    proizv = 1
    sum = 0
    for i in range(len(f_of[0])):
        if (i != 0):
            proizv *= ourX - arr_points[i - 1]
        sum += f_of[0][i] * proizv
    return sum
```

```python
find_functions()
print("Значение в точке X* = ", our_X, ", полученное многочленом Ньютона, равно  ", newton(funct_n, our_X, arr_points))
```

```python
x_newton = []
y_newton = []
for num in range(800):
    x_newton.append(-3 + num * 0.01)
    y_newton.append(newton(funct_n, -3 + num * 0.01, arr_points))
```

```python
plt.title('Newton')
plt.grid(True)
plt.scatter(math.pi / 4, newton(funct_n, our_X, arr_points), color='purple')
plt.plot(x_newton, y_newton, color='orange')
plt.plot(x_cos, y_cos, color='green')
plt.show()
```

```python
##############################
def diff_lagranj(ourX, arr_points):
    x = symbols('x')
    a = []
    
    for number in range(len(arr_points)):
        sum = 1
        for i in range(len(arr_points)):
            if (i != number):
                sum *= (x - arr_points[i]) / (arr_points[number] - arr_points[i])   
        a.append(diff(math.cos(arr_points[number]) * sum, x))
    sum = 0
    
    for number in range(len(a)):
        sum += a[number]
        
    return sum.subs({x:ourX})

#дифференциал Лагр. и функции
x_lagr_dif = []
y_lagr_dif = []

x_cos_diff = []
y_cos_diff = []
for num in range(800):
    x_lagr_dif.append(-3 + num * 0.01)
    y_lagr_dif.append(diff_lagranj(-3 + num * 0.01, arr_points))
    x_cos_diff.append(-3 + num * 0.01)
    y_cos_diff.append(-math.sin(-3 + num*0.01))

import matplotlib.pyplot as plt
plt.title('differentiation')
plt.grid(True)
plt.plot(x_lagr_dif, y_lagr_dif, color='blue')
plt.plot(x_cos_diff, y_cos_diff, color='green')
plt.show()
```
